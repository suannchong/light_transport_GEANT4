```python
import numpy as np
import uproot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(font_scale=2)
import os
```

### Useful functions

```python
def generate_data(folder, filename, array, increment):
    offset = 1/increment
    i,j,k = np.array(filename.split('_')[2:5]).astype(float)
    i,j,k = int(i/increment), int(j/increment), int(k/increment)
#     print(abs(j-offset),i+offset,abs(k-offset))
    counts = []
    file = uproot.open(folder + filename)
    events = file['Hits']
    energy = events.array("m_energy")
    position = events.array("m_position")
    x, y = position.x, position.y
    array[abs(k-offset)][abs(j-offset)][i+offset] = len(energy)
#     print(array)
    return x, y, energy

def get_entries(folder, filename):
    file = uproot.open(folder + filename)
    events = file['Hits']
    energy = events.array("m_energy")
    return len(energy)
```

# Investigate the distribution of light when optical photons are generated at a different points within the scintillator volume

```python
folder = '/Users/suannchong/Desktop/Geant4/mpnd_suann/build/results/'
directory = os.fsencode(folder)

x_total = []
y_total = []
energy_total = []
array = np.zeros((9,9,9))
ToT = []

for file in sorted(os.listdir(directory)):
    filename = os.fsdecode(file)
    if filename.endswith(".root"):
#         print(filename)
#         print(get_entries(filename))
        ToT.append(get_entries(folder, filename))
        x, y, energy = generate_data(folder,filename,array, 0.25)
        x_total.append(x)
        y_total.append(y)
        energy_total.append(energy)
        
    else:
        continue
```

```python
ToT_values = [i/10000 for i in ToT]
# print(ToT_values)
plt.hist(ToT_values,bins=np.arange(0.0, 0.2,0.01) )
```
